```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df = pd.read_csv("matches.csv")
df.head()        # first 5 rows
df.info()        # column details
df.describe()    # statistics
```

```python
df.dtypes
```

```python
df['season'].value_counts().plot(kind='bar')
plt.title("Matches per Season")
plt.xlabel("Season")
plt.ylabel("Number of Matches")
plt.show()
```

```python
sns.countplot(x='team1', data=df)
plt.xticks(rotation=90)
plt.title("Team1 Match Count")
plt.show()
```

```python
corr = df.corr(numeric_only=True)

sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

```python
top_teams = df['winner'].value_counts().head(5)

sns.barplot(x=top_teams.index, y=top_teams.values)
plt.title("Top Winning Teams")
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='winner', data=df, order=df['winner'].value_counts().index)
plt.xticks(rotation=90)
plt.title("Top Winning Teams")
plt.show()
```
