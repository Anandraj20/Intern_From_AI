```python
import matplotlib.pyplot as plt

x=[1,2,3,4]
y=[10,20,30,40]

plt.plot(x,y)

plt.xlabel("x values")
plt.ylabel("y values")
plt.title("Simple line graph")

plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import numpy as np

data=np.random.rand(10,10)
sns.heatmap(data,annot=True)

plt.show()
```

```python

import pandas as pd

data=pd.DataFrame({
    "Category" : ["A","A","B","B"],
    "Value" : [10,15,20,25],
    "Type" : ["X","Y","X","Y"]
})

sns.barplot(x="Category",y="Value", hue="Type" , data=data ,palette="bright")
plt.show()
```
