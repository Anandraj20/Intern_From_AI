```python
passs = ["abc123", "Hello@123", "weakpass"]
for ch in passs :
   has_digit=any(c.isdigit() for c in ch)
   has_sp=any(not c.isalnum() for c in ch)
   if(len(ch)>=8 and has_digit and has_sp) :
       print("Strong password")
   else :
       print("Weak password")
```

```python
word=input("Enter the string")
vcount=0
ccount=0
for ch in word :
    
    if(ch.lower()=="a" or ch.lower() == "i" or ch.lower()=="o" or ch.lower()=="e" or ch.lower()=="u") :
        vcount+=1
    else :
        ccount+=1
print(f"Vowels count{vcount}")
print(f"consonant count{ccount}")
```
