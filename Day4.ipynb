```python
import pandas as pd
import numpy as np

# Dataset 1 (Employee details)
df1 = pd.DataFrame({
    "EmpID": [1, 2, 3, 4],
    "Name": ["Anu", "Bala", None, "Divya"],
    "Age": [23, np.nan, 25, 22]
})

# Dataset 2 (Salary details)
df2 = pd.DataFrame({
    "EmpID": [1, 2, 3, 5],
    "Salary": [50000, 60000, np.nan, 70000]
})

print(df1)
print(df2)
```
