```python
#1. Total sales by product
df.groupby('Category')['Final_Price(Rs.)'].sum()
```

```python
#Average sales by region
df.groupby('Category')['Final_Price(Rs.)'].mean()
```

```python
#3. Number of payment methods by cateory
df.groupby('Category')['Payment_Method'].sum()
```

```python
#Total profit by region and product
df.groupby(['Price (Rs.)','Final_Price(Rs.)'])['Discount (%)'].sum()
```

```python
#Using Multiple Functions

#You can use .agg() to apply many functions at once

df.groupby('final_price(rs.)')['price (rs.)'].agg(['sum', 'mean', 'max'])
```

```python
df.groupby('final_price(rs.)')['price (rs.)'].sum().plot(kind='bar', title='Price by Category')
```

```python
df.groupby('Category')['Payment_Method'].value_counts().unstack().plot(kind='bar', title='Payment Method by Category')
```

```python
import matplotlib.pyplot as plt

df.columns = df.columns.str.strip().str.lower()

ax = df.groupby('category')['price (rs.)'].sum().plot(
    kind='bar', color='skyblue', label='Sales'
)

df.groupby('category')['price (rs.)'].sum().plot(
    kind='line', color='red', marker='o',
    secondary_y=True, ax=ax, label='price (rs.)'
)

ax.set_title('Price vs Profit by Product')
plt.show()
```
