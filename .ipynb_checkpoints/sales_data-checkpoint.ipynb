```python
#Fixing Data Types

# sure each column has the correct type.
```

```python
df1['Quantity'] = df1['Quantity'].astype(int)
```

```python
df1.info()
```

```python
df1['Date'] = pd.to_datetime(df1['Date'])
```

```python
df1.info()
```

```python
df1['Region'] = df1['Region'].str.strip().str.title()   # “south ” → “South”
df1['Product'] = df1['Product'].str.replace(r'[^A-Za-z0-9 ]', '', regex=True)
```

```python
df1
```

```python
#Feature Engineering (Creating New Columns)

#You can create new insights from existing data:
```

```python
df1['profit_margin'] = (df1['Profit'] / df1['Price']) * 100
df1['month'] = df1['Date'].dt.month_name()
df1['year'] = df1['Date'].dt.year
df1['total_amount'] = df1['Price'] * df1['Quantity']
```

```python
df1
```

```python
df1.sort_values(by='Quantity',ascending=False)
```

```python
df1=df1.rename(columns={'month':'Month'})
```

```python
df1.groupby('Region')['Price'].sum().plot(kind='bar', title='Price by Region')
```

```python
df1.groupby('Product')['Price'].sum().plot(kind='bar', color='skyblue', label='Sales')
df1.groupby('Product')['Profit'].sum().plot(kind='line', color='red', label='Profit', secondary_y=True)
plt.title('Price vs Profit by Product')
plt.legend()
plt.show()
```

```python
#price vs proft by product using seaborn


grouped_df = df1.groupby('Product')[['Price', 'Profit']].sum().reset_index()

# Create a figure
plt.figure(figsize=(10,6))

# Create barplot for Sales
sns.barplot(x='Product', y='Price', data=grouped_df, color='skyblue', label='Price')

# Create lineplot for Profit on secondary y-axis
ax1 = plt.gca()  # Get current axis
ax2 = ax1.twinx()  # Create second y-axis
sns.lineplot(x='Product', y='Profit', data=grouped_df, color='red', marker='o', label='Profit', ax=ax2)

# Add titles and labels
ax1.set_title('Price vs Profit by Product', fontsize=14)
ax1.set_xlabel('Product')
ax1.set_ylabel('Price', color='blue')
ax2.set_ylabel('Profit', color='red')

# Combine legends from both plots
lines1, labels1 = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines1 + lines2, labels1 + labels2, loc='upper left')

# Show the chart
plt.show()
```

```python
#price vs profit by product

# Group data by Product
grouped_df = df1.groupby('Product')[['Price', 'Profit']].sum().reset_index()

# Create figure
fig, ax1 = plt.subplots(figsize=(10,6))

# Bar chart for Price
bars = ax1.bar(grouped_df['Product'], grouped_df['Price'], color='skyblue', label='Price')

# Line chart for Profit
ax2 = ax1.twinx()
line = ax2.plot(grouped_df['Product'], grouped_df['Profit'], color='red', marker='o', label='Profit')

# Titles and labels
ax1.set_title('Price vs Profit by Product', fontsize=14, fontweight='bold')
ax1.set_xlabel('Product', fontsize=12)
ax1.set_ylabel('Price', color='blue', fontsize=12)
ax2.set_ylabel('Profit', color='red', fontsize=12)

# Add value labels on bars
for bar in bars:
    height = bar.get_height()
    ax1.text(bar.get_x() + bar.get_width()/2, height + 5, f'{int(height)}', ha='center', va='bottom', fontsize=9)

# Combine legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc='upper left')

# Add grid
ax1.grid(axis='y', linestyle='--', alpha=0.6)

# Show plot
plt.tight_layout()
plt.show()
```
