```python
import pandas as pd
import numpy as np

# Employees dataset (df1)
df1 = pd.DataFrame({
    'EmpID': [1, 2, 3, 4, 5],
    'Name': ['Anand', 'Raj', None, 'Kumar', None],
    'Age': [25, np.nan, 30, 28, np.nan],
    'Department': ['IT', 'HR', None, 'Finance', 'IT']
})

# Salary dataset (df2)
df2 = pd.DataFrame({
    'EmpID': [1, 2, 3, 6],
    'Salary': [50000, np.nan, 60000, 70000],
    'Bonus': [5000, None, None, 7000]
})
```
