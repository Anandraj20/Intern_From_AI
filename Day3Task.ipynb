```python
#Find Length of Each Word using map
word=["Anand","Raj","MohanRaj","Lokesh","Manikandan"]
res=list(map(len,word))
print(res)
```

```python
word=["Anand","Raj","MohanRaj","Lokesh","Manikandan"]
res=list(map(lambda x : len(x),word))
print(res)
```

```python
#Add Two Lists using map
a=[1,2,3]
b=[4,5,6]

res=list(map(lambda x,y: x+y ,a,b))
print(res)
```

```python
#2 .Convert Strings to Uppercase using map

s="anandraj"
s="".join(map(lambda x : x.upper(),s))
print(s)
```

```python
#4Check All Numbers are Positive using all

a=[1,2,-3,4,5,6]
print(all(x < 0 for x in a))
```

```python
#Check All Strings Length > 3 using all and map

word=["Anand","Raj","MohanRaj","Lokesh","Manikandan"]
print(all(len(x)  for x in word))
```

```python
#Check All Strings Length > 3 using all and map

word=["Anand","Raj","MohanRaj","Lokesh","Man"]
anss=list(map(lambda x : len(x) > 3 , word))
print(anss)
```

```python
#Check if Any Number is Negative for any and map

a=[1,-1,2,4,5,6]

res=any(map(lambda x : x < 0 , a))
print(res)
```
